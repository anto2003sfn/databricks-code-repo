```python
from pyspark.sql import Row
from datetime import datetime
class logger:
    def log_event(app_name, status, message):
        log_entry = Row(
            app_name=app_name,
            timestamp=datetime.now(),
            status=status,
            message=message
            )
        log_df = spark.createDataFrame([log_entry])
    # Append to a persistent logging table
        log_df.write.mode("append").saveAsTable("job_audit_logs")

```

```python
logger.log_event("JSON_Processor", "FAILED", "Processed 10,000 records")
```
