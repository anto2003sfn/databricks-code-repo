## Data Munging
### - In Data engineering world, data munging is the process of transforming and mapping the raw data into required format, more specific to downstream purposes like analytics, reporting and ML engineering<br>
### - Rarely ready to use like Data siloes, Messy entries and Noise<br>
### - Stages of Munging<br>
- Discovery<br>
- Structring<br> 
- Cleaning<br>
- Enriching<br>
- Validating <br>
- Publishing<br>


## Munging vs. ETL: diff<br>
### - ETL refers formally an automated and dedicated pipeline, designed to move the data from source to dataware house<br>
### - Munging refers to the exploratroy and adhoc way of transformation, how the particular dataset can be made useful for the projects

```python
display(rawdf1.where("id in ('4000001','4000003')"))
```

```python
display(rawdf1.coalesce(1)
        # .where("id in ('4000003','4000001')")
        .orderBy(["id","age"],ascending=[True,False])
        .dropDuplicates(subset=["id"])
        )


dedupdf2 =rawdf1.coalesce(1).orderBy(["id","age"],ascending=[True,False]).dropDuplicates(subset=["id"])

dedupdf2.show()
```

```python
from pyspark.sql.functions import initcap
display(dedupdf2.withColumn("source",lit("Retail")).withColumn("prof",initcap("profession")))
```
