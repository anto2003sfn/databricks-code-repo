```python
%sql

select p.product_id,product_name,category,sum(total_amount)
-- c.customer_id,sum(quantity) as quantity, sum(price_per_unit) as price
 from retail.schema_retail.silver_purchase p 
 left join retail.schema_retail.silver_customer c on p.customer_id=c.customer_id
 left join retail.schema_retail.silver_products pr on pr.product_id=p.product_id
where pr.category='Dairy'
 group by 1,2,3
```

```python
c_df = spark.read.table("retail.schema_retail.silver_customer")
p_df = spark.read.table("retail.schema_retail.silver_purchase")
pr_df = spark.read.table("retail.schema_retail.silver_products")
```
