```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

def say_hello():
    print("Hello!")

# Manual wrapping
decorated_hello = my_decorator(say_hello)
decorated_hello()
```

```python
@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
def myfirst_decor(func):
    def wrapper():
        print("implementation is started...")
        func()
        print("implementation completed...")
    return wrapper

@myfirst_decor
def say_progress():
    print("implemntation in progress...")

say_progress()
```

```python
import time
from functools import wraps

def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.time()
        print("Start time : ",start_time)
        result = func(*args, **kwargs) # Execute the actual function
        print("args : ",args[0])
        end_time = time.time()
        print(f"Execution time: {end_time - start_time} seconds")
        print("End time : ",end_time)
        return result
    return wrapper

@timer
def heavy_computation(input):
    time.sleep(2) # Simulating work
    print("Computation finished!", input)

heavy_computation("Stracy !!!")
```

```python
from functools import wraps
import time

def audit_query(log_level="INFO"):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            print(f"[{log_level}] Executing query via {func.__name__}...")
            
            start_time = time.time()
            # Execute the actual function (which returns a result set)
            result = func(*args, **kwargs) 
            end_time = time.time()
            
            # Count records (Works for BigQuery RowIterator or List of Dicts)
            # We convert to a list to count, or use .total_rows if available
            try:
                record_count = getattr(result, 'total_rows', len(list(result)))
            except:
                record_count = "Unknown"

            duration = round(end_time - start_time, 2)
            print(f"[{log_level}] Success: {record_count} records retrieved in {duration}s")
            
            return result
        return wrapper
    return decorator
```

```python
@audit_query(log_level="DEBUG")
def heavy_computation(input):
    time.sleep(2) # Simulating work
    print("Computation finished!", input)

@audit_query(log_level="INFO")
def heavy_computation_1(input):
    time.sleep(2) # Simulating work
    print("Computation finished!", input)


@audit_query(log_level="ERROR")
def heavy_computation_2(input):
    time.sleep(2) # Simulating work
    print("Computation finished!", input)
    

heavy_computation("Hi hi")
heavy_computation_1("Say Hi")
heavy_computation_2("Say Hello")
```
