```python
rawdf1=spark.read.csv("/Volumes/telecom_catalog_assign/landing_zone/landing_vol/customer/custsmodified",header=False,inferSchema=True).toDF("id","firstname","lastname","age","profession")
rawdf1.show(20,False)
display(rawdf1.take(20))
display(rawdf1.sample(.1))
```

```python
rawdf1.where("id !='ten' and id<>''").show()
```

```python
leftdf = rawdf1.where("id in ('4000001','4000002')").distinct()
```

```python
leftdf.show()
```

```python
rightdf = rawdf1.where("id in ('4000003','4000002')").dropDuplicates(["id"])
rightdf.show()
```

```python
print("inner join")
inner_df = leftdf.join(rightdf,how="inner",on="id")
inner_df.show()
```

```python
print("inner join without how")
inner_df = leftdf.join(rightdf,on="id")
inner_df.show()
```

```python
print("inner join without how")
left_df = leftdf.join(rightdf,how="left", on="id")
left_df.show()


print("inner join without how")
left_df = leftdf.join(rightdf,how="leftouter", on="id")
left_df.show()


print("inner join without how")
left_df = leftdf.join(rightdf,how="left_outer", on="id")
left_df.show()


```

```python
print("right join")
right_df = leftdf.join(rightdf,how="right",on="id")
right_df.show()

print("right join")
right_df = leftdf.join(rightdf,how="right_outer",on="id")
right_df.show()


print("right join")
right_df = leftdf.join(rightdf,how="rightouter",on="id")
right_df.show()

```

```python

print("full join")
full_df = leftdf.join(rightdf,how="fullouter",on="id")
full_df.show()


print("full join")
full_df = leftdf.join(rightdf,how="full",on="id")
full_df.show()




print("full join")
full_df = leftdf.join(rightdf,how="full_outer",on="id")
full_df.show()

```

```python
print("left semi")
left_semi_df = leftdf.join(rightdf,how="left_semi", on="id")
left_semi_df.show()

# extract the all the from left table machting with right table by giving the left df schema
```

```python
print("left anti")
left_anti_df = leftdf.join(rightdf,how="left_anti", on="id")
left_anti_df.show()

# extract the all the from left table unmachting with right table by giving the left table schema
# it is equivalent to exsists in sql
# it is very faster
```

```python
leftdf.explain()
```

```python
left_semi_df.explain()
```

```python
left_anti_df.explain()
```

```python
data = [(1, 'pakkir', '2025-09-24',2),
(2, 'sachin', '2026-09-13',3)]

print(data)

df = spark.createDataFrame(data,schema=["eid","ename","doj","mgr_id"])
```

```python
df.show()
```

```python
# df.join(df, on=["eid", "mgr_id"], how="inner").show()
```

```python
# emp_id, emp_name, joining_date , manager_id  
# --> list all the employees whose joining date is earlier than their managers
```

```python
emp_df = df.alias("emp")
mgr_df = df.alias("mgr")
emp_df.show()
mgr_df.show()
```

```python
from pyspark.sql import functions as F

rs = emp_df.join \
(mgr_df,F.col("emp.mgr_id") == F.col("mgr.eid"),"inner"). \
    where("mgr.doj>emp.doj")
rs.show()
```

```python


from pyspark.sql import functions as F

result = emp_df.join(mgr_df, F.col("emp.mgr_id") == F.col("mgr.eid"), "inner").filter(F.col("emp.doj") < F.col("mgr.doj")) .select(
    F.col("emp.eid"),
    F.col("emp.ename"),
    F.col("emp.doj").alias("emp_joining_date"),
    F.col("mgr.ename").alias("manager_name"),
    F.col("mgr.doj").alias("manager_joining_date")
)
result.show()
```

```python
df.createOrReplaceTempView("emp")
df.createOrReplaceTempView("mgr")
sqldf = spark.sql("select e.eid as eid,e.ename as ename, e.doj as doj, m.eid as manager_eid,m.ename as manger_name, m.doj as manager_doj from emp as e inner join mgr as m on e.mgr_id=m.eid where e.doj<m.doj"
)
sqldf.show()
                
```
