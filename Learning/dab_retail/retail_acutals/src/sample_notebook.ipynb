```python
def is_valid_string(s):
    # Requirement 1: Must contain at least one 'a'
    if 'a' not in s:
        return False
    print('l',list(s))
    print('s',sorted(s))
    print('set',set(s))
    print('set',set(s).issubset({'a', 'b'}))


    # Requirement 2: All 'a's must come before 'b's
    # Sorting 'aaabb' results in 'aaabb' (valid)
    # Sorting 'ba' results in 'ab' (invalid because it changed)
    return list(s) == sorted(s) and set(s).issubset({'a', 'b'})

# Test cases
# print(f"'aaa': {is_valid_string('aaa')}")    # True
# print(f"'ba': {is_valid_string('ba')}")      # False
# print(f"'aab': {is_valid_string('aab')}")    # True
print(f"'aaabb': {is_valid_string('aaabb')}")# True
# print(f"'a': {is_valid_string('a')}")# True
# print(f"'a': {is_valid_string(' ')}")# True
```

```python
def is_balanced(s):
    # Dictionary to map closing brackets to their opening counterparts
    bracket_map = {')': '(', ']': '[', '}': '{'}
    stack = []

    for char in s:
        # If it's a closing bracket
        if char in bracket_map:
            print('s',stack.pop())
            # Pop the top element if stack isn't empty, else use a dummy value
            top_element = stack.pop() if stack else '#'
            print(top_element)
            print(bracket_map[char])
            
            # Check if the popped element matches the required opening bracket
            if bracket_map[char] != top_element:
                return False
        # If it's an opening bracket, push it to the stack
        elif char in ['(', '[', '{']:
            stack.append(char)
            print(stack.append(char))
            
    # If the stack is empty, all brackets were matched
    return len(stack) == 0

# Test Cases
print(f"'([])': {is_balanced('()')}")      # True
# print(f"'([)]': {is_balanced('([)]')}")      # False (Wrong order)
# print(f"'(())[': {is_balanced('(())[')}")    # False (Incomplete)
# print(f"'{[()]}': {is_balanced('{[()]}')}")  # Tru
```

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        return "Stack is empty"

    def peek(self):
        return self.stack[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.stack) == 0

# Usage
s = Stack()
s.push("A")
s.push("B")
print(s.pop()) # Returns "B"
print(s.pop()) # Returns "A"
print(s.pop()) # Returns "none"
```
