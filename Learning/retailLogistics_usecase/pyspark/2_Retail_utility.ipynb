```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructField,StructType, StringType,IntegerType
from word2number import w2n
from pyspark.sql.functions import udf, col
```

```python
def word_to_num(value):
    try:
        return int (value)
    except:
        print("input is related to string")
        try:
            return w2n.word_to_num(value.lower())
        except:
            return None
        

value = word_to_num("fifty nine")
print(value)


word_to_num_udf = udf(word_to_num,IntegerType())
print(word_to_num_udf)
```

```python
def read_csv(
    spark,path,header=True,inferSchema=True,sep=","
):
    return spark.read.option("header",header).option("inferSchema", inferSchema).option("sep",sep).csv(path)


def read_json(spark,path,mline=True):
    return spark.read.json(path,multiLine=mline,mode="PERMISSIVE")
```

```python
def read_file(spark,file_type,path, header=True,inferSchema=True,sep=","):
    if file_type =='csv':
        csv_df = read_csv(spark,path,header,inferSchema,sep)
        return csv_df
    elif file_type == "json":
        json_df = read_json(spark,path)
        return json_df
    else:
        raise Exception ("File type not supported")
    
```

```python
def merge_df(df1,df2,allowMissingColumns=True):
    return(df1.unionByName(df2, allowMissingColumns=allowMissingColumns))
           
```

```python
def write_file(df,mode="overwrite",format="delta",path=path):
    df.write.mode(mode).format(format).save(path)
    print("delta file created")

def write_table(df,mode="overwrite",tableName="dummy"):
    df.write.mode(mode).format("delta").saveAsTable(tableName)
```

```python
def join_df(df1,df2, how="inner",on="shipment_id"):
    return (df1.join(df2,how=how,on=on))
```
