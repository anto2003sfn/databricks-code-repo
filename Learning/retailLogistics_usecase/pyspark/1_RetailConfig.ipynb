```python
# importing SparkSession
from pyspark.sql.session import SparkSession
spark = SparkSession.builder.appName("retaillogistics").getOrCreate()

```

```python
dbutils.widgets.text("catalog","retail")
catalog = dbutils.widgets.get("catalog")
dbutils.widgets.text("schema","logistics")
schema = dbutils.widgets.get("schema")

print("catalog : ",catalog)
print("schema : ",schema)
```

```python
# print(f"{catalog}")
spark.sql(f"create catalog if not exists {catalog}")

# 
```

```python
spark.sql(f"create schema if not exists {catalog}.{schema}")
```

```python
spark.sql(f"create volume if not exists {catalog}.{schema}.datalake")
```

```python
spark.sql(f"create volume if not exists {catalog}.{schema}.bronze")
spark.sql(f"create volume if not exists {catalog}.{schema}.silver")
spark.sql(f"create volume if not exists {catalog}.{schema}.gold")
```

```python
src = f"/Volumes/{catalog}/{schema}"
datalake = f"{src}/datalake"
bronze = f"{src}/bronze"
silver = f"{src}/silver"
gold = f"{src}/gold"
print('src-',src)
print('delta',datalake)
print('bronze',bronze)
print('silver',silver)
print('gold',gold)
```

```python
import json
JSON_file = json.dumps(
    {
        "source": src,
        "datalake": datalake,
        "bronze": bronze,
        "silver": silver,
        "gold": gold
    }
)
dbutils.notebook.exit(JSON_file)
```
