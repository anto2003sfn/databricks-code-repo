```python
df = spark.sql("""WITH daily_revenue AS (
    SELECT '2026-01-01'::date AS sale_date, 100 AS revenue UNION ALL
    SELECT '2026-01-02', 150 UNION ALL
    SELECT '2026-01-03', 200 UNION ALL
    SELECT '2026-01-04', 100 UNION ALL
    SELECT '2026-01-05', 300 UNION ALL
    SELECT '2026-01-06', 250 UNION ALL
    SELECT '2026-01-07', 110 UNION ALL
    SELECT '2026-01-08', 500
)
SELECT 
    sale_date,
    revenue,
    ROUND(AVG(revenue) OVER (
        ORDER BY sale_date 
        ROWS BETWEEN 6 PRECEDING AND CURRENT ROW
    ), 2) AS rolling_avg_7_days
FROM daily_revenue""")
```
