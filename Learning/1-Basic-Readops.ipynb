````python
%sql
CREATE TABLE retail.schema_retail.users (
    UserID INT NOT NULL PRIMARY KEY,
    UserName STRING,
    Email STRING,
    SignUpDate DATE
);
````

````python
%sql
DROP TABLE IF EXISTS retail.schema_retail.products ;
CREATE TABLE retail.schema_retail.products (
    ProductID INT NOT NULL,
    ProductName STRING,
    Price DECIMAL(10,2),
    CategoryID INT
);
ALTER TABLE retail.schema_retail.products ADD CONSTRAINT products_pk PRIMARY KEY (ProductID);
-- ALTER TABLE retail.schema_retail.products DROP CONSTRAINT products_pk;
````

````python
%sql
CREATE TABLE retail.schema_retail.purchases (
    PurchaseID INT PRIMARY KEY,
    UserID INT,
    ProductID INT,
    PurchaseDate DATE,
    Quantity INT,
    FOREIGN KEY (UserID) REFERENCES retail.schema_retail.users(UserID),
    FOREIGN KEY (ProductID) REFERENCES retail.schema_retail.products(ProductID)
);
````
