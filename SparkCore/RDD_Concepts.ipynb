# RDD-> Resilient Distributed Datasets<br>
 - RDD is a spark's foundational and collection of core objects<br>
 - It is immutable and fault talrent, running in parallel and distributed collection of objects
 - RDDs can be operated thorugh transformation and actions 
 - Transforms ex : map, filter, flatmap...
 - Actions ex : collect and count
 - It is well suited for unstructured data
 - It does not support for query execution and optimization features like predicate pushdonw





# DataFrame
- It is a collection of distributed objects and organized into named columns otherwise called as schema'd RDD
- It is similar to tables in relational database
- Dataframes provide optimization using catalyst optimizer
- It supports both DSL and SQL running on cost based optimizer
- Less type safety as column names are not checking during the compilation





# Dataset
 - It is a combination of RDD and DataFrame API
 - It is a distributed collection of typed objects
 - Type safe, object oriented API
 - Sometimes slower for unstructured data compared RDD
 - It is not supported in Pyspark

# Transformation
- it is a kind of operations performed on top of RDD or Dataframe to create new RDD and Dataframe
- 

# Actions
- it is a set of functions, operating on top of RDD or Dataframe

# Lazy Evaluation
- Spark does not run the transformation immediately instead, it builds the lineage in the form of DAG
- An action being called on top of transformation, spark executes the transformation and optimize the execution plan before performing the computation
