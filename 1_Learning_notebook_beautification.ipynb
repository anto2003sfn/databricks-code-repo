##DataBricks learning
- databricks is the unified data analytics platform solution designed for DE, DS and MLE workflows
- it has built around 4 importanct components such workspace, datalake integration, compuate resources(cluster) and job scheduling
- Workspace provides the collaborative environment for data bricks, where user can access and organize the databricks asset like notebook, libraries,dashboards and experiment
- cluster maintains the workloads handling data processing, computational platform of databricks
- Data lake integration with S3, GCS, ADLS,BLOB
- Job scheduling wrt orchestration


## Magical Commands in databricks notebook
%md: allows you to write markdown text to design the notebook.<br>
%run: runs a Python file or a notebook.<br>
%sh: executes shell commands on the cluster edge/client node.<br>
%fs: allows you to interact with the Databricks file system (Datalake command (cloud storage s3/adls/gcs)).<br>
%sql: allows you to run Spark SQL/HQL queries.<br>
%python: switches the notebook context to Python.<br>
%pip: allows you to install Python packages.<br>
