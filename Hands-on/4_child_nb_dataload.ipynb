##### Load the spark session

```python
from pyspark.sql import SparkSession
def getSpark():
    sparkSession = SparkSession.builder.appName("Spark DataFrames").getOrCreate()
    return spark;
```

##### Read the csv file and write the data into data bricks table 

```python

spark = getSpark();
df = spark.read.csv("dbfs:///Volumes/workspace/db_schema_sales/usage_metrics/mobile_os_usage.csv",header=True).toDF ("date","mobile_ops_sys","percentile")
df.show(5)
drop_df = spark.sql("drop table if exists db_schema_sales.mobile_os_usage");
df.write.saveAsTable("db_schema_sales.mobile_os_usage")
```
