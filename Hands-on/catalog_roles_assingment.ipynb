```python
%sql
CREATE OR REPLACE FUNCTION retail.hr_schema.mask_pii_fn(column_value STRING)
RETURN SELECT
  CASE
    WHEN is_account_group_member('HR_Group') THEN column_value
    ELSE 'redacted'
  END;
```

```python
%sql
ALTER TABLE retail.schema_retail.customer_txn ALTER COLUMN customer_id SET MASK retail.hr_schema.mask_pii_fn;
```
